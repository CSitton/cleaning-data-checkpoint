## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
##Eliminating data before year 1973 
filter_years_small_df = small_df.loc[(small_df['Years active'] >='1973')
                                & (small_df['Years active'] <'2023')]

print(filter_years_small_df)
```

```python
filter_years_small_df.info()
filter_years_small_df.head()
```

```python
filter_years_mid_df = mid_df.loc[(mid_df['Years active'] >='1973')
                                & (mid_df['Years active'] <'2023')]

print(filter_years_mid_df)
```

```python
filter_years_mid_df.groupby('Name')[['Years active', 'Name']].head()
```

```python
filter_yrs_mod_df = mod_df.loc[(mod_df['Years active'] >='1973')
                                & (mod_df['Years active'] <'2023')]

print(filter_yrs_mod_df)
```

```python
filter_yrs_mod_df.info()
filter_yrs_mod_df.head()
```

```python
filter_yrs_large_df = large_df.loc[(large_df['Years active'] >='1973')
                                & (large_df['Years active'] <'2023')]

print(filter_yrs_large_df)
```

```python
filter_yrs_large_df.info()
```
